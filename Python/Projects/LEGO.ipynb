```python
def theme_by_year(year):
    # filter data from merged_sets_themes based on given year and then sorted by id
    sorted_df = merged_sets_themes[merged_sets_themes['year'] == year].sort_values(by = ['id']) 
    
    # drop duplicates based on theme id
    sorted_df = sorted_df.drop_duplicates(subset='id', keep="first")
    
    # drop extra columns which are not required
    sorted_df.drop(columns = ['set_num', 'name_sets', 'theme_id', 'num_parts', 'parent_id', 'year'], inplace = True)
    
    # display the with required indexing
    display(sorted_df.reset_index(drop = True))
### ENTER CODE HERE ###
```

- Create a function called `set_parts` that takes as input the set number (as a string) and version number (as an integer) and outputs a DataFrame with each part in the set.  
- The DataFrame should have the following columns/column names: `set_num`, `version`, `name_sets`, `year`, `name_themes`, `part_num`, `name_part`, `name_color`,`quantity`,`is_spare`.  
- The Data should be sorted by `part_num` and then `name_color`.  
- The index should go from 0 to n-1.

```python
# merging multiple data frames (minifigs, inventory_minifigs, inventories, sets and themes to find required results
# 1. merging merged_sets_themes with inventories on set_num
# 2. merging inventory_parts and parts on part_num
# 3. merging merged_iparts_parts and color left on color_id and right on id
# 4. merging merged_iparts_parts_colors with merged_sets_themes_inventories left on inventory_id and right on inventories_id

cols_to_display_for_set_parts = ['set_num', 'version', 'name_sets', 'year', 'name_themes', 'part_num', 'name_part', 'name_color','quantity','is_spare', 'num_parts']

merged_sets_themes_inventories = pd.merge(merged_sets_themes, inventories, on = 'set_num')
merged_sets_themes_inventories.rename(columns = {'id_y': 'inventories_id'}, inplace = True)

merged_iparts_parts = pd.merge(inventory_parts[['inventory_id', 'part_num', 'color_id', 'quantity', 'is_spare']], parts[['part_num', 'name']], on='part_num')
merged_iparts_parts_colors = pd.merge(merged_iparts_parts, colors[['id', 'name']], left_on='color_id', right_on='id')

# rename and drop columns
merged_iparts_parts_colors.rename(columns = {'name_x': 'name_part', 'name_y':'name_color'}, inplace=True)
merged_iparts_parts_colors.drop(columns = ['id'], inplace=True)


merged_iparts_parts_colors_sti = pd.merge(merged_iparts_parts_colors, merged_sets_themes_inventories[['set_num', 'version', 'name_sets', 'year', 'name_themes', 'inventories_id', 'num_parts']],
                                          left_on='inventory_id', right_on='inventories_id')[cols_to_display_for_set_parts]

merged_iparts_parts_colors_sti.head()
```

```python
def set_parts(set_number, version_number):
    # filter merged_iparts_parts_colors_sti on given set number and version number
    filtered_df = merged_iparts_parts_colors_sti[(merged_iparts_parts_colors_sti['set_num'] == set_number) & (merged_iparts_parts_colors_sti['version'] == version_number)]
    
    # sort dataframe by part_num and then name_color
    filtered_df.sort_values(by = ['part_num', 'name_color'], inplace = True)
    
    # return final result with indexing from 0
    return filtered_df.reset_index(drop = True)
### ENTER CODE HERE ###
```
